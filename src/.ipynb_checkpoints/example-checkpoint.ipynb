```python
def split_string(string):
    if string.startswith('Compra') or string.startswith('Venta'):
        split_row = string.split("@")
        return pd.Series({
            'Action': split_row[0].split()[0],
            'Number': split_row[0].split()[1], #'Name': ' '.join(split_row[0].split()[2:]),
            'Price': split_row[1].split()[0].replace(',', '.'),
            'Currency': split_row[1].split()[1]
        })
    else:
        return pd.Series({
            'Action': None,
            'Number': None, #'Name': None,
            'Price': None,
            'Currency': None
        })
```

```python
# Apply the function to each row of the dataframe
df[['Action', 'Number', 'Price', 'Currency']] = df['Descripción'].apply(split_string)
```

```python
df['Action'].replace('Compra', 'buy', inplace=True)
df['Action'].replace('Venta', 'sell', inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Fecha'] + ' ' + df['Hora'])
```

```python
def compute_gains(df: pd.DataFrame, shares_held: dict, sell_row: pd.Series) -> float:
    """
    Computes the gains or losses for a sell transaction using the FIFO method.

    Args:
        df: A pandas DataFrame with columns "Name", "Action", "Date", "Number", and "Price".
        shares_held: A dictionary with the running totals of shares held.
        sell_row: A pandas Series representing the sell transaction.

    Returns:
        The gains or losses for the transaction as a float.
    """
    shares_bought_list = []
    shares_sold = sell_row['Number']
    sale_price = sell_row['Price']
    gains = 0

    # Iterate over the shares held for the same stock in the order they were bought
    buy_rows = df.loc[(df['Name'] == sell_row['Name']) & (df['Action'] == 'buy') & (df['FIFOSharesRem']>0)].iterrows()
    costs_rows = df.loc[(df['Name'] == sell_row['Name']) & (df['Descripción'].str.contains('Coste'))]
    for i, buy_row in buy_rows:
        shares_bought = {}
        # If there are enough shares to cover the sale, calculate the gains/losses and update the shares held
        if buy_row['Number'] <= shares_sold :
            gains += (sale_price - buy_row['Price']) * buy_row['Number']
            shares_sold -= buy_row['Number']
            shares_held[buy_row['Name']] -= buy_row['Number']
            shares_bought['Number'] = buy_row['Number']
            shares_bought['Price'] = buy_row['Price']
            shares_bought_list.append(shares_bought)
            df.loc[i, 'FIFOSharesRem'] -= buy_row['Number']
            #locate additional costs 
            gains += costs_rows['Unnamed: 8'][costs_rows['Date'] == buy_row['Date']].sum()
            print('Buy costs: ', costs_rows['Unnamed: 8'][costs_rows['Date'] == buy_row['Date']].sum())
        # Otherwise, split the sale between the shares and break out of the loop
        else:
            gains += (sale_price - buy_row['Price']) * shares_sold
            shares_held[buy_row['Name']] -= shares_sold
            shares_bought['Number'] = shares_sold
            shares_bought['Price'] = buy_row['Price']
            shares_bought_list.append(shares_bought)
            gains += costs_rows['Unnamed: 8'][costs_rows['Date'] == buy_row['Date']].sum()
            print('Buy costs: ', costs_rows['Unnamed: 8'][costs_rows['Date'] == buy_row['Date']].sum())
            break
    
    gains += costs_rows['Unnamed: 8'][costs_rows['Date'] == sell_row['Date']].sum()
    print('Sell costs: ', costs_rows['Unnamed: 8'][costs_rows['Date'] == sell_row['Date']].sum())

    return gains, shares_bought_list, df
```

```python
def calculate_returns(df: pd.DataFrame) -> float:
    """
    Calculates the total returns of a stock using the FIFO method.

    Args:
        df: A pandas DataFrame with columns "Name", "Action", "Date", "Number", and "Price".

    Returns:
        The total returns of the portfolio as a float.
    """
    # Sort the dataframe by date in ascending order
    df = df.sort_values('Date')

    # Create a dictionary to store the running totals of shares held
    shares_held = {}
    shares_sold = {}

    # Initialize the running totals to 0
    for name in df['Name'].unique():
        shares_held[name] = 0

    gains = {}
    boughts = {}
    # Iterate over the rows in the dataframe
    df_u = df.copy()
    for index, row in df.iterrows():
        # If the action is a sell, compute the gains/losses using the FIFO method
        if row['Action'] == 'sell':
            # Put if so that do consider 2-month period between seel and next buy
            g, gains['Shares Bought' + str(row['Date'])], df_u = compute_gains(df_u, shares_held, row)
            gains[row['Name'] + ' ' + str(row['Number']) + ' ' + str(row['Date'])] = g
            if (g<0): #& 
                #print(df[(df['ISIN'] == row['ISIN']) & (df['Action']=='buy') & (df['Date'] >= row['Date']) & (df['Date'] <= row['Date'] + pd.DateOffset(months=2))])
                df_empty = df[(df['ISIN'] == row['ISIN']) & (df['Action']=='buy') & (df['Date'] >= row['Date']) & (df['Date'] <= row['Date'] + pd.DateOffset(months=2))].empty
                if df_empty == False:
                    print('No se puede compensar pérdida por regla 2 meses')
                
            
            #print(df_u[~pd.isnull(df['FIFOSharesRem'])])
            

    return gains
```
